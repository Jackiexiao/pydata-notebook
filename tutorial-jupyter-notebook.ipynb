    & i_t = \sigma (W_{xi} x_t + W_{hi}h_{t-1} + W_{ci}c_{t-1} + b_i) \\

    & c_t = f_t c_{t-1} + i_c tanh(W_{xc}x_t + W_{hc}h_{t-1} + b_c) \\
